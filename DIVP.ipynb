```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

from google.colab import drive
drive.mount('/content/drive')

im = cv2.imread("/content/Bhavya.jpg")
im_1 = cv2.resize(im,(200,250))
cv2_imshow(im_1)

b = cv2.cvtColor(im_1, cv2.COLOR_RGB2GRAY)
cv2_imshow(b)

C = cv2.cvtColor(im_1, cv2.COLOR_RGB2HSV)
cv2_imshow(C)

d = cv2.cvtColor(im_1, cv2.COLOR_RGB2YCrCb)
cv2_imshow(d)
```
